#  Array Indexing and slicing

- NUMPY INDEXING AND SLICING OF VECTOR
- NUMPY INDEXING AND SLICING OF MTRIX
- BOOLEAN INDEXING

```python
import numpy as np
```

**Numpy indexing and Slicing of vector**

```python
arr = np.arange(11,21)
arr
```

```python
arr[8] # indexing
```

```python
# Slicing
arr[1:5]
```

```python
arr[:6] # starts with zero index
```

```python
arr[1:] # end with last index
```

```python
arr_slice = arr[1::2] # start with 1 and 2 steps
arr_slice
```

```python
arr[2]
```

**Numpy indexing and Slicng of matrix**

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]
```

```python
# indexing
arr[1,3]
```

```python
arr[0,1]
```

```python
# Slicing
slice = arr[0:2,1:3]
slice
```

```python
slice1 = arr[3:,3:]
slice1
```

```python
arr[:,2]
```

**Boolean indexing**

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr %2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
