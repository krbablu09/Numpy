```python
import numpy as np
```

**Arithmetic Operation**

```python
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])
print(a1)
print(a2)
```

```python
arr2 = a1 + a2 # size of both array are same
arr2
```

```python
a1 - a2
```

```python
a1 * a2
```

```python
a1 / a2
```

```python
a1 // a2
```

```python
a1**a2
```

**Broadcasting**

```python
l = [10,20,30,40]
arr = np.array(l)
arr
```

```python
arr +10
```

```python
arr2 = np.arange(1,26).reshape(5,5)
arr2
```

```python
arr2 +10
```

**Matrix Operations**

```python
a = np.array([[1, 2],[3, 4]])
b = np.array([[5, 6],[7, 8]])
print(a)
print(b)
```

```python
a *b # its not correct
```

```python
a @ b # multiplication of matrix
```

```python
np.dot(a,b) # multiplication of matrix
```

```python
a.T
```

```python
b.T
```

**Advance Array Manipulation**

##### *Stacking*

```python
a = np.array([1,2,3,4]) # use vectors not use matrix
b = np.array([5,6,7,8]) 
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```

##### *Spliting array*

```python
c = np.arange(16).reshape(4,4)
c
```

```python
np.hsplit(c,2)
```

```python
a = np.vsplit(c,2)
```

```python
for i in a:
    print(i)
```
