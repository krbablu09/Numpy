**Valid Sudoku**

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],

    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],

    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
]) 
s
```

```python
b = np.sum(s,axis = 1)
for i in b:
    if i !=45:
        print("Sudoku is not valid")
        break
else: 
    print("Sudoku for rows it is valid")
```

```python
c = np.sum(s, axis = 0)
for i in c:
    if i !=45:
        print("Sudoku is not valid")
        break
else: 
    print("Sudoku for columns it is valid")
```

```python
# method 1
slice1 = s[:3,:3]
slice2 = s[3:6,:3]
slice3 = s[6:9,:3]
slice4 = s[:3,3:6]
slice5 = s[3:6,3:6]
slice6 = s[6:9,3:6]
slice7 = s[:3,6:9]
slice8 = s[3:6,6:9]
slice9 = s[6:9,6:9]
a = slice1.sum()
d = slice2.sum()
e = slice3.sum()
f = slice4.sum()
g = slice5.sum()
h = slice6.sum()
i = slice7.sum()
j = slice8.sum()
k = slice9.sum()
t = np.array([a,d,e,f,g,h,i,j,k])
r = t == 45
r

   
```

```python
# method 2
valid = True
for i in range(0,9,3):
    for j in range(0,9,3):
        n = s[i:i+3,j:j+3]
        m = n.sum()
        if m != 45:
            valid = False

if valid:
    print("Its a valid Sudoku")
else:
    print("Its not a valid Sudoku")
```
